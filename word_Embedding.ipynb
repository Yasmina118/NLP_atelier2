## Apply TF-IDF technics

```python
import math


# Term Frequency (TF)
def calculate_tf(words):
  tf = {}
  for word in words:
    if word in tf:
      tf[word] += 1
    else:
      tf[word] = 1
  return tf

# Inverse Document Frequency (IDF)
def calculate_idf(num_docs, word_counts):
  idf = {}
  for word, count in word_counts.items():
    idf[word] = math.log(num_docs / (1 + count))
  return idf

# TF-IDF
def calculate_tfidf(tf, idf):
  tfidf = {}
  for word, tf_value in tf.items():
    tfidf[word] = tf_value * idf[word]
  return tfidf
num_docs = 1 
# Calculate TF
tf = calculate_tf(words)

# Calculate IDF
idf = calculate_idf(num_docs, tf)

# Calculate TF-IDF
tfidf = calculate_tfidf(tf, idf)

# Print the top 10 most important words (adjust as needed)
sorted_tfidf = sorted(tfidf.items(), key=lambda x: x[1], reverse=True)
print("Top 10 most important words:")
for word, score in sorted_tfidf[:10]:
  print(f"{word}: {score:.4f}")
```

```python
import gensim.downloader as api
# load the pre-trained fastText model
fasttext_model = api.load("fasttext-wiki-news-subwords-300")
word_pairs = [('ابتكار', 'الابتكار'), ('عمل', 'الأعمال'), ('ناجحة', 'النجاح')]
for pair in word_pairs:
    similarity = fasttext_model.similarity(pair[0], pair[1])
    print(f"Similarity between '{pair[0]}' and '{pair[1]}' using Word2Vec: {similarity:.3f}")
```

## Tsne Algorithm

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt


# Extracting vectors for the chosen words
vectors = [embeddings_dict[word] for word in words if word in embeddings_dict]

# Labels for the words (for plotting)
labels = [word for word in words if word in embeddings_dict]

# t-SNE model
tsne_model = TSNE(n_components=2, random_state=23, perplexity=40, n_iter=300)
new_values = tsne_model.fit_transform(vectors)

# Extracting the x and y coordinates for plotting
x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])

# Plotting
plt.figure(figsize=(12, 12)) 
for i in range(len(x)):
    plt.scatter(x[i], y[i])
    plt.annotate(labels[i],
                 xy=(x[i], y[i]),
                 xytext=(5, 2),
                 textcoords='offset points',
                 ha='right',
                 va='bottom')
plt.show()
```
